```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Example DataFrame (replace with your actual DataFrame)
df = pd.DataFrame({
    'petal_length': [1.4, 1.3, 1.5, 1.4],
    'petal_width': [0.2, 0.2, 0.2, 0.2],
    'species': ['setosa', 'setosa', 'versicolor', 'versicolor']
})

# Define colors for each species
colors = {'setosa': 'r', 'versicolor': 'g', 'virginica': 'b'}  # Add more colors if you have more species

# Create a new column with numerical codes for colors
df['color'] = df['species'].map(colors)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with custom colors based on species
for species, color in colors.items():
    species_data = df[df['species'] == species]
    ax.scatter(species_data['petal_length'], species_data['petal_width'], c=color, label=species)

ax.set_xlabel('PetalLengthCm')
ax.set_ylabel('PetalWidthCm')
ax.set_zlabel('Species')

plt.title('3D Scatter Plot Example')
plt.legend()
plt.show()
```

```python
k_rng = range(1,10)
sse=[]

for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[[ 'petal_length', 'petal_width']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel("Sum of Squared errors")
plt.plot(k_rng, sse)
```

```python
km = KMeans(n_clusters=3,random_state=0,)
y_predicted = km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head(150)
```

```python
#accuracy measure
from sklearn.metrics import confusion_matrix

# Convert df.cluster to string if it contains numerical labels
df['cluster'] = df['cluster'].astype(str)

# Calculate confusion matrix
cm = confusion_matrix(df['species'], df['cluster'])
print(cm)
```

```python
true_labels = df.species
predicted_labels= df.cluster

cm = confusion_matrix(true_labels, predicted_labels)
class_labels = ['Setosa', 'versicolor', 'virginica']

# Plot confusion matrix
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels)
plt.yticks(tick_marks, class_labels)

# Fill matrix with values
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j, i, str(cm[i][j]), ha='center', va='center', color='white')

plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset
df = sns.load_dataset('iris')

# Count the number of each species
species_counts = df['species'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Iris Species')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Plot bar chart
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='species')
plt.title('Count of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Group by species and calculate mean values
species_means = df.groupby('species').mean()

# Plot line graph
plt.figure(figsize=(10, 6))
species_means.plot(kind='line', marker='o')
plt.title('Mean Measurements of Iris Species')
plt.xlabel('Species')
plt.ylabel('Measurements (cm)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
